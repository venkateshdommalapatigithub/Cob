# outliers in duration


```python
def ext_dur(duration):
    try:
        return int(''.join(filter(str.isdigit,duration)))
    except:
        return None
    
```

```python
netflix_dataset1['duration'].apply(ext_dur)
```

```python
Q1_year=netflix_dataset1['release_year'].quantile(0.25)
Q2_year=netflix_dataset1['release_year'].quantile(0.75)
IQR_year=Q2_year - Q1_year
lower_bound=Q1_year - 1.5 * IQR_year
upper_bound=Q2_year + 1.5 * IQR_year
```

```python
cleaned_dataset=netflix_dataset1[(netflix_dataset1['release_year']< lower_bound) |(netflix_dataset1['release_year']> upper_bound)]
cleaned_dataset
```
